```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load your data
a = np.load("C:\\Users\\SAI REVANTH\\Downloads\\activation_maps_binarization_Bangli-P9.jpg.npy")

# Assuming 'a' contains features, and we will use one feature as independent and the other as dependent

# Select two features (e.g., feature 0 as independent and feature 1 as dependent)
independent_feature = a[:, 0]
dependent_feature = a[:, 1]

# Create a Linear Regression model
model = LinearRegression()

# Reshape the independent feature to 2D (required for scikit-learn)
independent_feature = independent_feature.reshape(-1, 1)

# Fit the model
model.fit(independent_feature, dependent_feature)

# Make predictions
predicted_values = model.predict(independent_feature)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(dependent_feature, predicted_values)



# Print the Mean Squared Error
print(f'Mean Squared Error (MSE): {mse}')
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load your data for dataset 'b'
data_b = np.load("C:\\Users\\SAI REVANTH\\Downloads\\activation_maps_binarization_mask_Bangli-P9.jpg.npy")

# Assuming 'data_b' contains features, and we will use a single feature for linear regression

# Select the single feature for dataset 'b'
feature_b = data_b[:, 0]

# Create a Linear Regression model for dataset 'b'
model_b = LinearRegression()
feature_b = feature_b.reshape(-1, 1)
model_b.fit(feature_b, feature_b)  # Predicting the same feature

# Make predictions
predicted_values_b = model_b.predict(feature_b)

# Calculate the Mean Squared Error (MSE)
mse_b = mean_squared_error(feature_b, predicted_values_b)

# Plot the original data and the linear regression line for dataset 'b'
plt.figure(figsize=(10, 6))
plt.scatter(feature_b, feature_b, label='Original Data (Dataset B)')
plt.plot(feature_b, predicted_values_b, color='blue', label='Linear Regression Line')
plt.xlabel('Feature (Dataset B)')
plt.ylabel('Predicted Feature')
plt.title('Univariate Linear Regression (Dataset B)')
plt.legend()

# Print the Mean Squared Error for dataset 'b'
print(f'Mean Squared Error (MSE) for Dataset B: {mse_b}')

plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load your data for the dataset with continuous values
data = np.load("C:\\Users\\SAI REVANTH\\Downloads\\activation_maps_binarization_Bangli-P9.jpg.npy")

# Assuming 'data' contains features and continuous target values

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data[:, :-1], data[:, -1], test_size=0.2, random_state=42)

# Create and train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions using the model
y_pred = model.predict(X_test)

# Calculate Mean Squared Error (MSE) to evaluate the model
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Load your data
a = np.load("C:\\Users\\SAI REVANTH\\Downloads\\activation_maps_binarization_Bangli-P9.jpg.npy")
b = np.load("C:\\Users\\SAI REVANTH\\Downloads\\activation_maps_binarization_mask_Bangli-P9.jpg.npy")

# Assuming 'a' contains features and 'b' contains labels (target)

# Select two features (e.g., feature 0 and feature 1)
feature1 = a[:, 0]
feature2 = a[:, 1]

# Combine the features
X = np.column_stack((feature1, feature2))
y = b  # Target values

# Create a Regression Tree model
tree_regressor = DecisionTreeRegressor(random_state=0)
tree_regressor.fit(X, y)

# Make predictions using the model
tree_regressor.predict(X)

```
